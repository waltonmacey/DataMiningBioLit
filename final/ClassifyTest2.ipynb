```python
import re
import csv

csvread = csv.reader(open('trainingdata.csv', 'rb'), delimiter=',', quotechar='"')
sent = []
featureList = []




for row in csvread:
    sentence = row[0]
    typ = row[1]
    featureList.append(sentence)
    sent.append((sentence, typ));
    
wordList = []
for sentstr in featureList:
    word = re.sub("[^\w]", " ",  sentstr).split()
    wordList.extend(word)
```

```python
def extract_features(sentence):
    s_words = sentence.split()
    features = {}
    for word in wordList:
        features['contains(%s)' % word] = (word in s_words)
    return features
```

```python


wordList = list(set(wordList))

training_set = nltk.classify.util.apply_features(extract_features, sent)


train_data = training_set[:90]
test_data = training_set[90:]
```

```python
model = nltk.NaiveBayesClassifier.train(training_set)

test1 = 'THis is all about data and 4 people had 5 mg, 56 = 4.589'
test2 = 'The results we did discover are all shown here'
print model.classify(extract_features(test1))
```

```python
print 'accuracy:', nltk.classify.util.accuracy(model, test_data)
```

def word_feats(words):
    return dict([(word, True) for word in words])

negfeats = [(word_feats(movie_reviews.words(fileids=[f])), 'neg') for f in negids]
posfeats = [(word_feats(movie_reviews.words(fileids=[f])), 'pos') for f in posids]
 
negcutoff = len(negfeats)*3/4
